## Estandarización del Archivo .Json

<p style="text-align: justify;">Teniendo en cuenta el material de Samir (2015), se debe transformar el conjunto de datos, especialmente en la columna '_id', para que contenga solamente solo datos de tipo numérico, ya que es el indice conjunto de datos.</p>

```python
#se procede a crear el DataFrame a partir del JSON
df = pd.DataFrame(df)

"""se crea una función, con su respectivo condicional para manejar los valores de
   id que al verificar tenían estructura de diccionario"""
def handle_dict_id(value):
    if isinstance(value, dict):

        return str(value)
    else:
        return value

"""se utiliza el metodo .apply()para aplicar una función a 
cada elemento de una serie o columna en un DataFrame.columna '_id'"""
df['_id'] = df['_id'].apply(handle_dict_id)

#se procede convertir la columna '_id' de variable categórica a numérica, empezando el indice de 1 a N
df['_id'] = pd.Categorical(df['_id']).codes + 1# se suma un 1 para que no empiece el indice desde cero (0)

#se procede con guardar el rarchivo estandarizado nombrandolo con otro nombre
df.to_json('output/libraries_standarizated.json', orient='records')

df
```
